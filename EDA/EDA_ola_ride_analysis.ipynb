#### Ride Volume Over Time

```python
# 1. Ride Volume Over Time
df['Date'] = pd.to_datetime(df['Date'])
ride_volume = df.groupby('Date')['Booking_ID'].count().nlargest(20)  # top 20 busiest days

ride_volume.plot(kind='bar', figsize=(12,6), color='green')
plt.title("Top 20 Days by Ride Volume")
plt.xlabel("Date")
plt.ylabel("Number of Rides")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##### 1. Why did you pick the specific chart?

- **Effective Comparison:** A bar chart is one of the most effective ways to compare discrete categories (in this case, dates) against a single metric (number of rides).  

- **Clear Axes Representation:**  
  - **X-axis:** Shows the top 20 dates (time series data).  
  - **Y-axis:** Represents ride counts — making it easy to visually compare which dates had the highest ride volumes.  

- **Visual Hierarchy:** The bar lengths provide a clear visual hierarchy.  
  - Longer bars = higher ride volume  
  - Shorter bars = lower ride volume  

##### 2. What is/are the insight(s) found from the chart?

- **Consistent Ride Volume:** Most of the top days have either **9 or 10 rides**, showing stable demand.  

- **Multiple Peaks:** Several dates tie for the highest ride volume (**10 rides**), meaning there are multiple peak periods rather than a single standout day.  

- **No Major Spikes or Drops:** The ride counts don’t show extreme highs or lows — demand is **fairly evenly distributed** across these top days.  

- **Spread Across the Month:** The top 20 days are spread throughout **July 2024**, suggesting no single week or event solely drove the high activity.

#### Heatmap of correlations

```python
# 18. Heatmap of correlations
plt.figure(figsize=(10,7))

# Drop unwanted columns before correlation
corr = df.drop(columns=['Canceled_Rides_by_Customer', 'Canceled_Rides_by_Driver'], errors='ignore') \
         .corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Without Canceled Ride Columns)")
plt.show()
```

##### 1. Why did you pick the specific chart?



- **Correlation Heatmap:** A correlation heatmap is an excellent way to visually represent a correlation matrix.  
- **Color Coding:** Different colors indicate the strength and direction of relationships between pairs of variables:  
  - **Red:** Strong positive correlation  
  - **Blue:** Strong negative correlation  
- **Quick Visual Insight:** Makes it easy to spot correlations at a glance.

##### 2. What is/are the insight(s) found from the chart?



- **No Linear Relationships:** Almost all variables have a correlation of **0.00**, meaning there is no predictable linear relationship.  
  - For example, **Booking_Value** does not consistently increase or decrease with **Ride_Distance**.  
- **Diagonal Values:** The diagonal is all **1.00**, as expected, since a variable is perfectly correlated with itself.  
- **Independence of Metrics:** The very low correlation values suggest that metrics like **V_TAT, C_TAT, Booking_Value, Ride_Distance, Driver_Ratings, Customer_Rating** are largely independent of each other.
